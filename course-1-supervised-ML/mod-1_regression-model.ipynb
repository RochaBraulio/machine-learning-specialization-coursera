## Linear Regression with one variable
Goal is to use simple linear regression to predict the price of a house based on its area

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ucimlrepo
```

```python
# Create training dataset
# Source https://archive.ics.uci.edu/dataset/9/auto+mpg
auto_mpg = ucimlrepo.fetch_ucirepo(id=9) 
cars = pd.DataFrame(auto_mpg.data.features)
cars['mpg'] = auto_mpg.data.targets
```

```python
# Checking the first 5 rows of the dataset
cars.head(5)
```

```python
# Visualizing the relationship between horsepower and miles per gallon (mpg)
plt.figure(figsize=(10, 6))
plt.scatter(cars['horsepower'], cars['mpg'])
plt.ylabel('Miles per Gallon (mpg)')
plt.xlabel('Horsepower')
plt.title('Horsepower vs. MPG')
plt.show()
```
