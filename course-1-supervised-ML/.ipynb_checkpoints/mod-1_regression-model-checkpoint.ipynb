## Linear Regression with one variable
Goal is to use simple linear regression to predict the price of a house based on its area

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ucimlrepo
import time
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
```

```python
# Create training dataset
# Source https://archive.ics.uci.edu/dataset/9/auto+mpg
auto_mpg = ucimlrepo.fetch_ucirepo(id=9) 
cars = pd.DataFrame(auto_mpg.data.features)
cars['mpg'] = auto_mpg.data.targets
```

```python
# Checking the first 5 rows of the dataset
cars.head(5)
```

```python
# Visualizing the relationship between horsepower and miles per gallon (mpg)
plt.figure(figsize=(10, 6))
plt.scatter(cars['horsepower'], cars['mpg'])
plt.ylabel('Miles per Gallon (mpg)')
plt.xlabel('Horsepower')
plt.title('Horsepower vs. MPG')
plt.show()
```

```python
X = np.array(cars['horsepower'])
y = np.array(cars['mpg'])
```

# Example from the book

```python
X = 2*np.random.rand(100,1)
y = 8 + 2*X + np.random.rand(100,1)
print(X.ndim,X.shape)
```

```python
plt.scatter(X,y)
plt.axis([0, 2, 8, 13])
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta_best)
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 8, 13])
plt.show()
```

```python
def create_training_X_set(m):
    X = 2*np.random.rand(m,1)
    return X
    
def create_training_Y_set(m):
    y = 8 + 2*X + np.random.rand(m,1)
    return y
    
def solve_normal_eq(X, y, m):
    X_b = np.c_[np.ones((m,1)), X]
    theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    return theta_best
```

```python
test_rng = [10**i for i in range (1,9)]
test_perf = []
mod_sol = []
t0 = 0
dt = 0

for test in test_rng:
    X = create_training_X_set(test)
    y = create_training_Y_set(test)
    t0 = time.perf_counter()
    sol = solve_normal_eq(X, y, test)
    dt = (time.perf_counter()-t0)*1000
    mod_sol.append(sol)
    test_perf.append(dt)

print(test_rng)
print(test_perf)
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(test_rng, test_perf, color='blue', alpha=0.7, label='Data Points', s=50)
ax.set_xscale('log')
ax.set_xlabel('Instances in training set (Log Scale)', fontsize=12)
ax.set_ylabel('Time in miliseconds', fontsize=12)
ax.grid(True, which="both", ls="-", alpha=0.3) # 'which="both"' ensures grid lines for major and minor ticks
plt.tight_layout()
plt.show()
```
